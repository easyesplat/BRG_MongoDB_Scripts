```python
import numpy as np 
import pandas as pd

pd.set_option('display.max_columns', None)
```

### Hurricane Michael

```python
michael_building_data = pd.read_csv("../Data/Hurricane_Michael/HM_D2D_Building.csv")
michael_building_data.head(50)
```

```python
michael_building_data.info()
```

```python
print("% of photos: " + str(1 - (michael_building_data['all_photos'].isna().sum() / len(michael_building_data))))

note_columns = michael_building_data.filter(like='note')
print("% of descriptions: " + str(1 - (note_columns.isna().all(axis=1).sum() / len(michael_building_data))))
```
